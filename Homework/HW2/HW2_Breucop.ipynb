```python
import random
import pandas as pd

from sklearn.cross_validation import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.cross_validation import KFold

import warnings
warnings.filterwarnings("ignore")
```

```python
with open('./iris.csv', 'rb') as f:
    irises = pd.read_csv(f, header = None, index_col = False)
irises
```

```python
irises.columns = ['sepal length','sepal width','petal length','petal width','class']
irises.dtypes
```

```python
X = irises[['sepal length','sepal width','petal length','petal width']]
y = irises['class']
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.20, random_state=0)
```

```python
myknn = KNeighborsClassifier(3).fit(X_train,y_train)
```

```python
myknn.score(X_test,y_test)
```

This exercise largely mimicked the lab work. Loading the data was perhaps the most difficult part but the .dropna() method worked pretty swimmingly. And now we'll cross validate with 5 folds. (Reminder for self: each fold uses 1 set as the validation set and the other 4 as training). For a bit of a challenge I've tried forgoing the KFold package.

```python
def cross_validate(X, y, classifier, k_fold) :

    # derive a set of (random) training and testing indices
    k_fold_indices = KFold( len(X), n_folds=k_fold,
                           indices=True, shuffle=True,
                           random_state=0)

    k_score_total = 0
    # for each training and testing slices run the classifier, and score the results
    for train_slice, test_slice in k_fold_indices :

        model = classifier(X.iloc[ train_slice  ],
                         y.iloc[ train_slice  ])

        k_score = model.score(X.iloc[ test_slice ],
                              y.iloc[ test_slice ])

        k_score_total += k_score

    # return the average accuracy
    return k_score_total/k_fold
```

```python
cross_validate(X,y,KNeighborsClassifier(120).fit,5)
```

```python
k_list = [cross_validate(X,y,KNeighborsClassifier(K).fit,5) for K in xrange(3,120) ]
#note that 3 is the 0th index
bi = k_list.index(max(k_list)) + 3
print bi,max(k_list)
```

```python
import matplotlib.pyplot as plt
    
%matplotlib inline
```

```python
plt.plot(xrange(3,120),k_list, 'bs')
plt.title("Accuracy vs. Number of Neighbors")
plt.show
```

```python
plt.plot(xrange(50,90),
         [ cross_validate(X,y,KNeighborsClassifier(K).fit,5)*100 for K in xrange(50,90) ]
         , 'ro')
```

This precipitous drop in accuracy appears right around when two thirds of the training set has been established, dropping down to about 55%. From a frequentist perspective, slightly better than flipping a coin suggests that at this high of a number of nighbors, our model is better at determining one group the new data point does not belong to but has a random chance of determining between the other two. This also explains the drop to about 33% (1/3) towards the end, when the system might as well make a guess.

```python
plt.plot(xrange(2,150),
         [ cross_validate(X,y,KNeighborsClassifier(11).fit,K)*100 for K in xrange(2,150) ]
         , 'ro')
```

Banding! The point of cross validation is to determine how good your classifier is so I believe a truly optimal number of folds isn't reasonable in an exact sense, however, numbers towards the low in make sense as the values appear to converge around 96.7% in this case. Choosing an "optimal" value as in a minimum or maximum wouldn't make sense since you'd intentionally skew your model better or worse arbitrarily.
