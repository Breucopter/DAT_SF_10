```python
import pandas as pd
import numpy as np
import datetime as dt

import matplotlib.pyplot as plt

%matplotlib inline
```

```python
pd.set_option('display.max_rows',100)
pd.set_option('display.max_columns',60)
```

```python
data = pd.read_csv('credit_analysis.csv', header=None)
```

```python
data.head()
```

```python
data.info()
```

```python
data.replace({17 : { '-' : 0, '+' : 1}},inplace=True)
```

```python
type([x for x in set([x for x in data[5]])][0])
```

```python
for x in data.columns:
    if data[x].dtypes == "O":
        new = str(x) + '_?'
        data.replace({x : {'?':new}},inplace=True)
    else:
        continue
```

```python
data[16] = data[16].apply(lambda x: (dt.datetime.now() - dt.datetime.strptime(x, '%Y-%m-%d')).days)
```

```python
# categorical_features = [1]
categorical = [1,4,5,6,7,9,10,12,13]
for n in categorical:
    temp_df = pd.get_dummies(data[n])
    
    data.drop(n,axis=1,inplace=True)
    data = pd.merge(data, temp_df,left_index=True,right_index=True)
```

```python
data.head()
```

```python
numeric_features = [2,14]
for n in numeric_features:
    data[n] = data[n].convert_objects(convert_numeric=True,convert_dates=True)
```

```python
data.dropna(inplace=True)
```

```python
target = data[17]
```

```python
from sklearn.cross_validation import KFold

def cross_validate(X, y, classifier, k_fold) :

    # derive a set of (random) training and testing indices
    k_fold_indices = KFold(len(X), n_folds=k_fold,
                           shuffle=True, random_state=0)

    k_score_total = 0
    # for each training and testing slices run the classifier, and score the results
    for train_slice, test_slice in k_fold_indices :

        model = classifier(X[ train_slice  ],
                         y[ train_slice  ])

        k_score = model.score(X[ test_slice ],
                              y[ test_slice ])

        k_score_total += k_score

    # return the average accuracy
    return k_score_total/k_fold
```

```python
# subset.head()
data_norm = (data - data.mean())/data.std()
data_norm.drop(['4_?','5_?'],axis=1,inplace=True)
data_norm.head()

subset_cols = [x for x in data_norm.columns if x != 17]
subset = data[subset_cols]
subset_norm = data_norm[subset_cols]
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(C=1).fit
subset_norm.head()
```

```python
fitted_lr = model(subset,target)
x = np.arange(len(subset.columns))
plt.bar(x, fitted_lr.coef_.ravel())
_ = plt.xticks(x + 0.5, subset.columns, rotation=30)
```

```python
fitted_lr = model(subset_norm,target)
x = np.arange(len(subset_norm.columns))
plt.bar(x, fitted_lr.coef_.ravel())
_ = plt.xticks(x + 0.5, subset_norm.columns, rotation=30)
```

The first column looks promising! But it isn't: systematic bias based on the fact that yes comes toward the end and no at the beginning. It is an index column and a red herring.

```python
subset_norm.drop(0,axis=1,inplace=True)
fitted_lr = model(subset_norm,target)
x = np.arange(len(subset_norm.columns))
plt.bar(x, fitted_lr.coef_.ravel())
_ = plt.xticks(x + 0.5, subset_norm.columns, rotation=30)
```

```python
cross_validate(subset_norm.values,target.values,model,10)
# print len(subset_norm),len(target)
```

```python
from sklearn import svm

clf = svm.LinearSVC().fit
cross_validate(subset_norm.values,target.values,clf,10)
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(C=1).fit
cross_validate(subset_norm.values,target.values,lr,10)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(max_depth=10).fit

cross_validate(subset_norm.values,target.values,rf,10)
```
