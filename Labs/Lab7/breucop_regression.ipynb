Yay for an order of magnitude improvement! Let's check in with cross-validation.

```python
from sklearn.cross_validation import KFold
def cross_validate(X, y, classifier, k_fold) :

    # derive a set of (random) training and testing indices
    k_fold_indices = KFold( len(X), n_folds=k_fold,
                           indices=True, shuffle=True,
                           random_state=0)

    k_score_total = 0
    # for each training and testing slices run the classifier, and score the results
    for train_slice, test_slice in k_fold_indices :

        model = classifier(X.iloc[ train_slice  ],
                         y.iloc[ train_slice  ])

        k_score = model.score(X.iloc[ test_slice ],
                              y.iloc[ test_slice ])

        k_score_total += k_score

    # return the average accuracy
    return k_score_total/k_fold
```

So with this I used PCA on pieces but I haven't normalized the data. Let's try to get more out of this with a global normalization and then dimensionality reduction.

```python
logins_v4 = logins[['account_age','problems_completed','student_previous_logins_total','student_previous_class_logins','a','c','g','e','m']]
```

```python
logins_v4.head()
```

```python
logins_n = pd.DataFrame(columns = logins_v4.keys())

for key in logins_v4.keys():
    v4_max = logins_v4[key].max()
    v4_min = logins_v4[key].min()
    v4_range = v4_max-v4_min
    v4_mean = logins_v4[key].mean()
    logins_n[key] = logins_v4[key].map(lambda z: (z - v4_mean)/v4_range)
    
```

```python
global_pca = PCA(n_components=1)
global_transform = activity_pca.fit_transform(logins_n)
logins_n['global_pca'] = global_transform

print "Max: ",  np.max(global_transform, axis=0)
print "Min: ", np.min(global_transform,axis = 0)
print "Mean: ", np.mean(global_transform, axis = 0)
print "Variance: ", np.var(global_transform,axis = 0)
logins_n.head(10)
```

```python
logins_g = pd.DataFrame(columns=['global_pca'])
logins_g['global_pca'] = logins_n['global_pca']
```

```python
g_train, g_test, g_y_train, g_y_test = train_test_split( logins_g, y, test_size=0.20, random_state=0)
global_PCA_regr = LinearRegression().fit(g_train,g_y_train)
global_PCA_regr.score(g_test,g_y_test)
```

```python
print "Large Featureset Model: ", cross_validate(logins_v2, y, LinearRegression().fit,5)
print "Model w/ PCA: ",cross_validate(logins_v3, y, LinearRegression().fit,5)
print "Model w/ Normalized PCA: ",cross_validate(logins_g, y, LinearRegression().fit,5)
```
